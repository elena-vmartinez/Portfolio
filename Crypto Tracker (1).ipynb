# Automation

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time

def automated_crypto_import():

    url = 'https://coinmarketcap.com/currencies/ethereum/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    name0 = soup.find('span', class_='sc-f70bb44c-0 jltoa', title='Ethereum').text.strip()
    name = name0.replace(' price', '')

    price0 = soup.find('span', class_='sc-f70bb44c-0 jxpCgO base-text').text
    price = price0.replace('$','')

    date_time = datetime.now()

    dict = {'Crypto Name': name,
           'Price': price,
           'TimeStamp': date_time}

    df = pd.DataFrame([dict])

    if os.path.exists('/Users/Elena/Downloads/Crypto_Tracker/Ethereum.csv'):
        df.to_csv('/Users/Elena/Downloads/Crypto_Tracker/Ethereum.csv', mode = 'a', 
                  header = False, index = False)
    else:
        df.to_csv('/Users/Elena/Downloads/Crypto_Tracker/Ethereum.csv', index=False)
        
    print(df)
```

```python
while True:
    automated_crypto_import()
    time.sleep(3600)
```
